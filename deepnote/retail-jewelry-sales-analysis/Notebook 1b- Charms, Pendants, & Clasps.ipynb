```
enriched = pd.read_csv('/work/charmenriched.csv', encoding = 'latin-1')

enriched.describe()
```

```
df = enriched

# --- Compute summary by Type ---
profit_summary = (
    df.groupby('Type', dropna=False)
      .agg({
          'Retail Profit (num)': 'mean',
          'Retail Margin (%)': 'mean'
      })
      .reset_index()
      .sort_values('Retail Profit (num)', ascending=False)
)

# --- Bar chart: Retail Profit ---
fig_profit = px.bar(
    profit_summary,
    x='Type',
    y='Retail Profit (num)',
    text='Retail Profit (num)',
    title='Average Retail Profit by Type',
    color='Retail Profit (num)',
    color_continuous_scale='Blues'
)
fig_profit.update_traces(
    texttemplate='$%{text:,.0f}',
    textposition='outside'
)
fig_profit.update_layout(
    xaxis_title='Type',
    yaxis_title='Average Retail Profit ($)',
    template='simple_white',
    paper_bgcolor='#EAEAEA',
    plot_bgcolor='#EAEAEA',
    height=600
)
fig_profit.show()

# --- Bar chart: Retail Margin (%) ---
fig_margin = px.bar(
    profit_summary,
    x='Type',
    y='Retail Margin (%)',
    text='Retail Margin (%)',
    title='Average Retail Margin (%) by Type',
    color='Retail Margin (%)',
    color_continuous_scale='Tealgrn'
)
fig_margin.update_traces(
    texttemplate='%{text:.1f}%',
    textposition='outside'
)
fig_margin.update_layout(
    xaxis_title='Type',
    yaxis_title='Average Retail Margin (%)',
    template='simple_white',
    paper_bgcolor='#EAEAEA',
    plot_bgcolor='#EAEAEA',
    height=600
)
fig_margin.show()
```

```
# --- Compute totals by Type ---
cost_structure = (
    df.groupby('Type', dropna=False)
      .agg({
          'Total Cost (num)': 'mean',
          'Retail Price (num)': 'mean',
          'Retail Profit (num)': 'mean'
      })
      .reset_index()
      .sort_values('Retail Price (num)', ascending=False)
)

# Melt for grouped bar format
cost_melted = cost_structure.melt(
    id_vars='Type',
    value_vars=['Total Cost (num)', 'Retail Price (num)', 'Retail Profit (num)'],
    var_name='Metric',
    value_name='Amount'
)

# --- Plotly grouped bar ---
fig = px.bar(
    cost_melted,
    x='Type',
    y='Amount',
    color='Metric',
    barmode='group',
    text='Amount',
    title='Average Cost, Price, and Profit by Type',
    color_discrete_map={
        'Total Cost (num)': '#E69F00',     # orange
        'Retail Price (num)': '#56B4E9',   # blue
        'Retail Profit (num)': '#009E73'   # green
    }
)

fig.update_traces(
    texttemplate='$%{text:,.0f}',
    textposition='outside'
)

fig.update_layout(
    xaxis_title='Type',
    yaxis_title='Average Amount ($)',
    template='simple_white',
    height=600,
    legend_title_text='',
    paper_bgcolor='#EAEAEA',
    plot_bgcolor='#EAEAEA',
)

fig.show()
```
